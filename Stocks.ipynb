```python
#get stock data with ticker
ticker = 'KTOS'
stock = web.DataReader(ticker, "av-daily-adjusted", start=datetime(2015,1, 1), end=datetime(2021, 1, 16),api_key=os.getenv('ALPHAVANTAGE_API_KEY'))
```

```python
#compute the daily returns
stock['pct_change'] = stock['adjusted close'].pct_change()
stock['log_ret'] = np.log(stock['adjusted close']) - np.log(stock['adjusted close'].shift(1))
stock['rolling_ma_20'] = stock['adjusted close'].rolling(window=20).mean()
stock
```

```python
#chart the stock price across time
stock.close.plot()
stock.rolling_ma_20.plot()
```
