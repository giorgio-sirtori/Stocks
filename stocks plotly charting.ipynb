```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_datareader.data as web
from datetime import datetime
import cvxopt as opt
from cvxopt import blas, solvers
import os
%matplotlib inline
%env ALPHAVANTAGE_API_KEY = NTN3D6P6LIMZ6A2Q#alpha advantage api key
```

```python
ptf = ['PLTR','KTOS','BABA','PDD']
ptf_df = pd.DataFrame()
```

```python
for stock in ptf:
    ptf_df[stock] = web.DataReader(stock, "av-daily-adjusted", 
                                        start=datetime(2015,1, 1), 
                                        end=datetime(2021, 1, 16),api_key=os.getenv('ALPHAVANTAGE_API_KEY'))['adjusted close']
```

```python
ptf_df
```

```python
#returns
returns_df = pd.DataFrame()
for each in ptf_df.columns:
    returns_df[each] = ptf_df[each].pct_change()
```

```python
returns_df
```

```python
import plotly.graph_objects as go
import plotly.io as pio

fig = go.Figure(go.Scatter(y=returns_df['PLTR'],x=returns_df.index, text = returns_df.index))
fig.update_layout(title_text='PLTR Returns')
#pio.write_html(fig, file='hello_world.html', auto_open=True)
```

```python
import plotly.express as px

fig = px.bar(returns_df, x=returns_df.index, y="PLTR")
fig.show()
```

```python
fig = px.line(ptf_df, x=ptf_df.index, y=ptf_df.columns)
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")
fig.show()
```
